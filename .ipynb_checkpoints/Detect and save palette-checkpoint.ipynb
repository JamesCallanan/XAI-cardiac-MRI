```python
def map_colors(im, palette, palette_ref):
    im_new = np.zeros(np.shape(im))
    for pix_x in range(np.shape(im)[0]):
        for pix_y in range(np.shape(im)[1]):
            for col, col_ref in zip(palette,palette_ref):
                z = np.argmin(np.linalg.norm(im[pix_x,pix_y,:]-col))
                ref = np.argmin(np.linalg.norm(im[pix_x,pix_y,:]-col_ref))        
                im_new[pix_x,pix_y,0] = col[z] - col[ref] - im[pix_x,pix_y,0]
                im_new[pix_x,pix_y,1] = col[z] - col[ref] - im[pix_x,pix_y,1]
                im_new[pix_x,pix_y,2] = col[z] - col[ref] - im[pix_x,pix_y,2]
    return im_new
    
#im_new = map_colors(im1, palette2, palette1)
```

```python
def to_rgb(src, pal):
    cols = set(src.flatten())
    shp = np.shape(src)
    if len(shp) != 3:
        tm = np.zeros((shp[0],shp[1],3))
        for i in range(1,5):
            tm[:,:,0] = np.where(src==i, pal[i-1][0], tm[:,:,0])
            tm[:,:,1] = np.where(src==i, pal[i-1][1], tm[:,:,1])        
            tm[:,:,2] = np.where(src==i, pal[i-1][2],tm[:,:,2])    
    return tm.astype('int')
```
